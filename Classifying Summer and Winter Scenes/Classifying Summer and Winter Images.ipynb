#### Format of the image files
Each file begins with a line containing two numbers. The product of these two numbers equals the number of
lines in each file (minues the first one). This suggests the rows represent elements in a 2-dimensional grid. 

On the first line, the first of the two numbers is the height of the image (in pixels) and the second is the width (again in pixels). Each line in the rest of the file contains the pixels of the image. Each pixel is a triplet of numbers denoting how much red, green, and blue the pixel contains, respectively.

In image processing, each column in one of these image files is called a *channel* (disregarding line 1).  So there are 3 channels: red, green, and blue.

#### Reformat text files
First, we'll create a function called `read_file_lines` that takes in a filename as its argument.
This function should return a Python list containing the lines of the
file as strings. That is, if `1.txt` contains:

```
1 2 3
3 4 5
7 8 9
```

the return value should be: `['1 2 3\n', '3 4 5\n', '7 8 9\n']`.

Then, use the `read_file_lines` function on the file `1.txt`, reading the contents
into a variable called `file1`.

````python
def read_file_lines(filename):
    with open(filename, 'r') as f:
        return f.readlines()

file1 = read_file_lines('1.txt')
file1[:5] # See the first 5 lines
````

#### Convert to Image File
Now, we'll define a function called `lines_to_image` that takes in the contents of a
file as a list (such as `file1`). It should return an array containing integers of
shape `(n_rows, n_cols, 3)`. That is, it contains the pixel triplets organized in the
correct number of rows and columns.

For example, if the file originally contained:

```
4 2
0 0 0
10 10 10
2 2 2
3 3 3
4 4 4
5 5 5
6 6 6
7 7 7
```

The resulting array should be a *3-dimensional* array that looks like this:

```
array([
    [ [0,0,0], [10,10,10] ],
    [ [2,2,2], [3,3,3] ],
    [ [4,4,4], [5,5,5] ],
    [ [6,6,6], [7,7,7] ]
])
```

**Important note:** We must call `.astype(np.uint8)` on the final array before
returning so that `numpy` will recognize the array represents an image.

````python
def lines_to_image(file_lines):
    rows, cols = [int(num) for num in file_lines[0].split()]
    stripped_lines = [line.strip() for line in file_lines[1:]]
    triplets = [[int(num) for num in line.split()]
                for line in stripped_lines]
    triplet_arr = np.array(triplets)
    image_array = triplet_arr.reshape((rows, cols, 3))
    return image_array.astype(np.uint8)

image1 = lines_to_image(file1)
print(image1.shape)
image1[:2] # Show first 2 lines of image file
````

#### Show Image

Images in `numpy` are simply arrays, but we can also display them them as
actual images in this notebook.

Now, we'll use the `show_images` function defined below to display the first image, `image1`. 

````python
def show_images(images, ncols=2, figsize=(10, 7), **kwargs):
    """
    Shows one or more color images.
    
    images: Image or list of images.  Each image is a 3-dimensional
            array, where dimension 1 indexes height and dimension 2
            the width.  Dimension 3 indexes the 3 color values red,
            blue, and green (so it always has length 3).
    """
    def show_image(image, axis=plt):
        plt.imshow(image, **kwargs)
        
    if not (isinstance(images, list) or isinstance(images, tuple)):
        images = [images]
    images = [image.astype(np.uint8) for image in images]
    
    nrows = math.ceil(len(images) / ncols)
    ncols = min(len(images), ncols)
    
    plt.figure(figsize=figsize)
    for i, image in enumerate(images):
        axis = plt.subplot2grid(
            (nrows, ncols),
            (i // ncols,  i % ncols),
        )
        axis.tick_params(bottom=False, left=False, top=False, right=False,
                         labelleft=False, labelbottom=False)
        axis.grid(False)
        show_image(image, axis)
````

````python
# Show image1 here:
show_images(image1)
````

#### Rescale Pixel Values 

Unfortunately, the picture above is completely black. If you look at the data, you'll notice all the numbers lie between 0 and 10. In `NumPy`, a color intensity is an integer ranging from 0 to 255, where 0 is no color (black).  That's why the image is almost black.  To see the image, we'll need to rescale the numbers in the data to have a larger range.

Because of this, we'll define a function `expand_image_range` that takes in an image and returns a
**new copy** of the image with the following transformation:
    
    old value | new value
    ========= | =========
    0         | 12
    1         | 37
    2         | 65
    3         | 89
    4         | 114
    5         | 137
    6         | 162
    7         | 187
    8         | 214
    9         | 240
    10        | 250

This expands the color range of the image. For example, a pixel that previously had the value `[5 5 5]` (almost-black) will now have the value `[137 137 137]` (gray).

````python
transformed = np.array([12, 37, 65, 89, 114, 137, 162, 187, 214, 240, 250])

def expand_image_range(image):
    return transformed[image]

expanded1 = expand_image_range(image1)
show_images(expanded1)
````

#### Show All Images

Now that we're able to see the image after rescaling the pixels, we can now define a function called `reveal_file` that takes in a filename and returns an expanded image. 

Then, we'll set `expanded_images` to a list of all the revealed images. Finally, we'll use `show_images` to display the `expanded_images`.

````python
def reveal_file(filename):
    return expand_image_range(lines_to_image(read_file_lines(filename)))

filenames = ['1.txt', '2.txt', '3.txt', '4.txt', '5.txt','6.txt', '7.txt', '8.txt', '9.txt', '10.txt']
expanded_images = [reveal_file(filename) for filename in filenames]
show_images(expanded_images, ncols=5)
````

Now that we've seen the image files, we now need to think about how we'd distinguish between pictures of summer and winter. One trait that seems specific to summer pictures is that the colors are warmer.

Let's see if the proportion of pixels of each color in the image can let us distinguish between summer and winter pictures.

#### Calculate Pixel Proportions
To simplify things, we can categorize each pixel according to its most intense (highest-value) channel. (Remember, red, green, and blue are the 3 channels.) For example, we could just call a `[2 4 0]` pixel "green." If a pixel has a tie between several channels, let's count it as none of them.

Let's write a function `proportion_by_channel`. It takes in an image.  It assigns each pixel to its greatest-intensity channel: red, green, or blue. Then the function returns an array of length three containing the proportion of pixels categorized as red, the proportion categorized as green, and the proportion categorized as blue (respectively). (Again, don't count pixels that are tied between 2 or 3 colors as any category, but do count them in the denominator when you're computing proportions.)

For example:

```
>>> test_im = np.array([
  [ [5, 2, 2], [2, 5, 10] ] 
])
>>> proportion_by_channel(test_im)
array([ 0.5, 0, 0.5 ])

# If tied, count neither as the highest
>>> test_im = np.array([
  [ [5, 2, 5], [2, 50, 50] ] 
])
>>> proportion_by_channel(test_im)
array([ 0, 0, 0 ])
```

Then, set `image_proportions` to the result of `proportion_by_channel` called
on each image in `expanded_images` as a 2d array.

````python
def proportion_by_channel(image):
    
    NUM_CHANNELS = 3
    
    # Get total number of pixels from image size
    n_pixels = image.shape[0] * image.shape[1]
    
    # Flatten image array using total number of pixels
    flattened = image.reshape((n_pixels, NUM_CHANNELS))
    
    # Sort each pixel from low to high within itself
    sorted_by_channel_value = np.sort(flattened, axis=1)
    
    # Find indices of the pixels where 2nd position  does not equal the 3rd position (tied for most intense)
    indices_with_winner = sorted_by_channel_value[:,NUM_CHANNELS-1] != sorted_by_channel_value[:,NUM_CHANNELS-2]
    
    # Get the pixels where there is a "winner"
    pixels_with_winner = flattened[indices_with_winner,:]
    
    # Count the number of occurences for each pixel
    counts = np.bincount(np.argmax(pixels_with_winner, axis=1), minlength=NUM_CHANNELS)
    return counts / n_pixels

image_proportions = np.array([proportion_by_channel(image) for image in expanded_images])
image_proportions
````

Let's plot the proportions computed above on a bar chart:

````python
pixel_proportions = pd.DataFrame({
        'red': image_proportions[:, 0],
        'green': image_proportions[:, 1],
        'blue': image_proportions[:, 2]
    }, index=pd.Series(['Image {}'.format(n) for n in range(1, 11)], name='image'))\
    .iloc[::-1].plot.barh(color = ['r', 'g', 'b']);
````

#### Define the Classification Algorithm

Reviewing the images above, the summer images were images 2, 3, 5, 6, 9. Looking at the color proportions of those images, it looks like the summer images have a much larger proportion of red pixels compared to the winter images. Knowing this, we can set a red proportion threshold of 0.25 to determine whether or not the image is summer or winter. If the red proportion threshold is greater than 0.25, then the image will be classified as summer. If not, it will be classified as winter. Let's define a function that does this.

````python
RED_PROP_THRESHOLD = 0.25

def summer_or_winter(image):
    if proportion_by_channel(image)[0] > RED_PROP_THRESHOLD:
        return "Summer"
    else:
        return "Winter"

[summer_or_winter(image) for image in expanded_images]
````

````python
image_scenes['Prediction'] = [summer_or_winter(image) for image in expanded_images]
image_scenes['Accurate?'] = image_scenes['Scene'] == image_scenes['Prediction']
image_scenes
````
